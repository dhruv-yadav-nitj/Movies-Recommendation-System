<h1> Data Preparation
<h4> This dataset needed very much treatement since it was not at all in the needed format

```python
movies = pd.read_csv('/content/gdrive/MyDrive/movie_recommender/tmdb_5000_movies.csv')
credits=pd.read_csv('/content/gdrive/MyDrive/movie_recommender/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
# since we have two datasets -> we need to merge them into one (to make it easy)
print(movies.shape, credits.shape)
movies = movies.merge(credits, on='title')
print(movies.shape)
```

```python
movies.head()
```

```python
'''
Useful Columns in the Movies Dataset:
-> genres -> keywords -> title -> overview -> cast -> crew
'''
```

```python
selected_columns= ['genres', 'title', 'id', 'keywords', 'overview',  'cast', 'crew']
movies = movies[selected_columns]

''' another method for same thing:
movies = movies[['genres', 'title', 'id', 'keywords', 'overview',  'cast', 'crew']]
'''
print(movies.shape); movies.head()
```

```python
# here we would be slightly pre-processing the dataset
'''
what are we doing in pre-processing:
-> removing those movies which has any of the columns missing (NULL)
-> removing any movies that is duplicated
-> treating of the data format in columns, genre format, cast (top 3) and crew (only 1 (director)) (for easeness)
'''

# removing null values
print(movies.isnull().sum()); movies.dropna(inplace=True); print(movies.shape)

# removing duplicates values
print(movies.duplicated().sum()) # since none data was duplicate -> no action

movies.iloc[0]['genres']
# this line would be giving output as this:
''' [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] '''
# but it would not be feasible for the algorithm to work on this -> so we would be bringing this format to like -> ['Action', 'Adventure'....]

```

```python
from ast import literal_eval # converts string to list format (required here since the given genre is in the string type)
```

```python
# Preprocessing Pt.3 (treating genre column format)
def genre (obj):
    genre_list = list()
    for i in literal_eval(obj):
        genre_list.append(i['name'])
    return genre_list

movies['genres'] = movies['genres'].apply(genre)
```

```python
# Preprocessing Pt. 4 (doing the same thing for keyword column as we did for genre column)
movies['keywords'] = movies['keywords'].apply(genre)
```

```python
print(movies.iloc[0]['cast'])
```

```python
# Preprocessing Pt 5 (selecting top 3 casts for each movie)
def cast (obj):
    counter, cast_list = 0, list()
    for i in literal_eval(obj):
        cast_list.append(i['name'])
        counter += 1
        if counter == 3:
            break
    return cast_list
movies['cast'] = movies['cast'].apply(cast)
```

```python
# Preprocessing Pt 6 (selecting top 1 crew (the director) for each movie)
def cast (obj):
    crew_list = list()
    for i in literal_eval(obj):
        if i['job'] == 'Director':
            crew_list.append(i['name'])
            break
    return crew_list
movies['crew'] = movies['crew'].apply(cast)
```

```python
# Preprocessing Pt 7 (converting the overview string into a list)
movies['overview'] = movies['overview'].apply (lambda x : x.split(' '))
```

```python
# Preprocessing Pt 8
# removing spaces between the same element strings (why necessary)
# suppose we are searching for movies of Sam Worthington and there is another person names as Sam Mendis, so until there is space between Sam and Worthington, the model will recognise the two words different so it would start recommending Sam Mendis movies with Sam Worthington movies
# for more clear explanation 50th minute

movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ", "") for i in x])

movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['cast'] + movies['crew'] + movies['keywords'] + movies['overview']
movies.head()
```

```python
new_columns = ['id', 'title', 'tags']
new_df = movies[new_columns] ; new_df.head()
```

```python
# convert the list ('tag') into string
new_df['tags'] = new_df['tags'].apply (lambda x: " ".join(x)) ; new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply (lambda x : x.lower()); new_df.head()
```

```python
# Preprocessing Pt. 10 -> performing stemming/lemmatization -> we would be reducing words to their base form. eg. 'activity' and activities' should be only 'activity'
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
def do_stem (obj):
    hehe = list()
    for i in obj.split():
        hehe.append(stemmer.stem(i))
    return " ".join(hehe)

new_df['tags'] = new_df['tags'].apply(do_stem)
new_df.head()
```

<h1> Modelling

**Text vectorization** is a crucial step in natural language processing (NLP) and machine learning tasks involving text data. It involves converting textual data into a numerical format that machine learning algorithms can understand. Here are some common methods for text vectorization:

***Bag of Words (BoW):***
Represents a document as an unordered set of words, ignoring grammar and word order but keeping track of word frequency.
Each word in the document is assigned a unique index, and the vector represents the count of each word in the document.

***Term Frequency-Inverse Document Frequency (TF-IDF):***
Similar to Bag of Words but takes into account the importance of words by considering both the frequency of a term in a document and its rarity across all documents in the dataset.
Helps to downweight common words and highlight important ones.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# vectorization-function
cv_function = CountVectorizer(stop_words='english', max_features=5000)
# converting into numpy-array
vectors = cv_function.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# we would be calculating cosine_distance of each movie with all other movie (means vector)
similarity = cosine_similarity(vectors)
print(similarity.shape)
```

```python
def recommend (movie):
    movie_index = new_df[new_df['title'] == movie].index[0] # fetching index of movie
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]

    recommended_movies_index = list()
    for i in movies_list:
        recommended_movies_index.append(i[0])
    return recommended_movies_index
```

<h1> Testing & Validation

```python
search_for = "The Conjuring"
print("Search For:", search_for)

try:
    ans = recommend(search_for)
    for i in ans:
        print(new_df.iloc[i]['title'])
except:
    print("EITHER THE MOVIE DOES NOT EXIST OR THE NAME IS NOT CORRECT !")
```
